1. What is the difference between a function and a method in Python ?

Ans :    

a. Function:

* A function is a standalone block of code that is not bound to any object or class.
* It is defined using the def keyword and can take input arguments, perform operations, and return a result.
* Functions can be called from anywhere in the code, as long as they are in scope.

Example

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result)
```

b. Method:

* A method is a function that is defined inside a class, making it associated with that class or its instances (objects).
* Methods are called on an instance of a class, and they have access to the instance’s data.
* The first parameter of a method is usually self, which refers to the instance of the class calling the method. This allows methods to access and modify the object's attributes.

Example


```python
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(3, 5)
print(result)
```

2. Explain the concept of function arguments and parameters in python.

Ans :    

a. Parameters

Parameters are the names defined in a function's signature (definition) that specify what kind of input the function can accept. They act as placeholders for the values that will be provided when the function is called. Parameters are variables defined between the parentheses in the function definition.

Types of Parameters in Function
* Positional Parameters: Parameters without default values. They are mandatory.
* Default Parameters: Parameters with default values. Optional to provide arguments.
* Keyword-Only Parameters: Defined after * in the function definition, must be passed using keyword arguments.
* Variable-Length Parameters: Use *args and **kwargs to allow a variable number of inputs.

Example of Parameters

```python
def greet(name, age):  # 'name' and 'age' are parameters
    print(f"Hello, {name}! You are {age} years old.")
```

b. Arguments

Arguments are the actual values or data you pass to a function when calling it. These values are assigned to the corresponding parameters in the function definition. In other words, arguments are the real data provided to a function’s parameters during a function call.



```python
greet("Abhishek", 25)  # "Abhishek" and 25 are arguments
```

Types of Function Arguments in Python

i) Positional Arguments
* These are the simplest form of arguments.
* The arguments are assigned to the parameters based on their position (order) in the function call.

```python
def add(a, b):
    return a + b

result = add(3, 5)  # 3 is assigned to 'a', 5 to 'b'
```

ii) Keyword Arguments
* These arguments are specified by the parameter name, allowing you to pass arguments in any order.
* This makes the code more readable and avoids confusion about the order of arguments.

```python
def greet(name, age):
    print(f"Hello, {name}! You are {age} years old.")

greet(age=25, name="Abhishek")
```

iii) Default Arguments
* You can define default values for parameters in the function.
* If the caller does not provide an argument for a parameter with a default value, the default value is used.

```python
def greet(name, message="Hello"):
    print(f"{message}, {name}!")

greet("Abhishek")
greet("Abhishek", "Hi")
```

vi) Variable-Length Arguments
* These allow you to pass a variable number of arguments to a function.
* Python provides * args for non-keyword variable arguments and  * * kwargs for keyword variable arguments.

Example of *args :     

```python
def add_all(*numbers):
    return sum(numbers)

result = add_all(1, 2, 3, 4)
```

Example of **kwargs :    

```python
def print_details(**details):  # Collects keyword arguments into a dictionary
    for key, value in details.items():
        print(f"{key}: {value}")

print_details(name="Abhishek", age=25, city="New Delhi")
```

4. What is the purpose of the return statement in a Python function ?

Ans :    

The return statement in a Python function is used to send back a result from the function to the caller. It allows a function to produce an output, which can then be captured and used elsewhere in the code. Without a return statement, a function will return None by default, indicating that it doesn't produce any meaningful output.

Purpose of the return Statement

a. Returning a Value:
* The primary purpose of the return statement is to provide a value back to the caller of the function. This value can be of any data type (integer, float, string, list, dictionary, object, etc.).

Example:

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result)
```

b. Ending Function Execution:
* The return statement immediately ends the function's execution. Any code after a return statement will not be executed.

Example:

```python
def check_number(num):
    if num > 0:
        return "Positive"
    return "Non-Positive"

print(check_number(10))
```

c. Returning Multiple Values:
* In Python, you can return multiple values from a function using tuples. This allows a function to provide more than one piece of information at once.

Example:

```python
def get_person_info():
    name = "Abhishek"
    age = 25
    city = "New Delhi"
    return name, age, city  # Returns a tuple with three values

info = get_person_info()
print(info)
```

d. Returning Nothing (None):
* If a function does not have a return statement or if the return statement is used without any value, it returns None.

Example:

```python
def greet(name):
    print(f"Hello, {name}!")

result = greet("Abhishek")  # Function does not return any value
print(result)
```

e. Returning Conditional Values:
* The return statement can be used to return different values based on conditions within the function.

Example:

```python
def is_even(num):
    if num % 2 == 0:
        return True
    else:
        return False

print(is_even(4))
print(is_even(5))
```

5. What are iterators in Python and how do they differ from iterables ?

Ans :    

Iterators
An iterator is an object that represents a stream of data. It is an object with a state so that it remembers where it is during iteration. Iterators have two main methods:

a. iter(): Returns the iterator object itself and is called when an iteration is initialized.

b. next(): Returns the next item in the sequence. If there are no more items, it raises a StopIteration exception.

Iterators are created from iterables using the iter() function, and you can manually retrieve the next item using the next() function.

Example:



```python
numbers = [1, 2, 3, 4, 5]
iterator = iter(numbers)  # Getting an iterator from the list

# Using the iterator to get items one by one
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

Key Differences Between Iterables and Iterators

a. Definition:
* Iterable: An object that can be iterated over. It must implement the __iter__() method, which returns an iterator.
* Iterator: An object that keeps track of the current position during iteration. It implements both the __iter__() and __next__() methods.

b. Usage:
* Iterable: You can use it with a for loop or pass it to functions that require an iterable (like sum() or min()).
* Iterator: Used to manually traverse items one by one using the next() function until a StopIteration exception is raised.

c. State:
* Iterable: Does not keep track of iteration state. It can generate a fresh iterator each time.
* Iterator: Keeps track of the iteration state (i.e., the current position in the sequence).

d. Creating Iterators:
* Iterable: You can create an iterator from an iterable using the iter() function.
* Iterator: Already an iterator; doesn't need conversion.

```python
# An iterable (list)
numbers = [10, 20, 30]

# Creating an iterator from the iterable
iterator = iter(numbers)

# Iterating using the iterator
print(next(iterator))
print(next(iterator))
print(next(iterator))

```

6. Explain the concept of generators in Python and how they are defined.

Ans :    

Generator

A generator is a function that returns an iterator object. Instead of using the return statement, a generator uses the yield statement to produce a series of values over time, pausing each time a value is yielded and resuming from that point when the generator is called again.

Key Concepts

* Lazy Evaluation: Generators compute values one at a time as needed. This is efficient in terms of memory, especially for large data sets.
* Statefulness: Generators maintain their state between calls. Each time the generator’s next() method is called, the function resumes execution right after the last yield statement.

Defining a Generator

You define a generator using a regular function with the yield keyword.

Example:

```python
def simple_generator():
    yield 1
    yield 2
    yield 3

gen = simple_generator()
print(next(gen))
print(next(gen))
print(next(gen))

```

How Generators Work
* yield 1 pauses the function and returns 1. The state of the generator is saved.
* The next time next() is called, it resumes from the point after yield 1, so it yields 2.
* This continues until there are no more values to yield.

Benefits of Using Generators

* Memory Efficient: No need to store the entire dataset in memory. Suitable for processing large data streams.
* Infinite Sequences: Generators can represent infinite sequences (e.g., reading from an infinite data source or an open-ended loop).
* Simple to Implement: Easier than implementing an iterator manually.

Example: Generator for Fibonacci Numbers


```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()
for _ in range(10):
    print(next(fib))

```

7. What are the advantages of using generators over regular function ?

Ans :    

Using generators in Python offers several advantages over regular functions, particularly when dealing with large datasets or complex data processing tasks. Here are the key benefits:

a. Memory Efficiency
* Generators do not store all values in memory. Instead, they generate each value on the fly when requested. This makes them extremely memory efficient, especially for large datasets or infinite sequences.
* Regular functions with large datasets would require you to create and store all data in memory at once, potentially leading to memory exhaustion.

Example:

```python
# Generator version
def generate_numbers():
    for i in range(1, 1000000):
        yield i

# Regular function version
def create_numbers():
    return [i for i in range(1, 1000000)]
```

b. Lazy Evaluation
* Generators use lazy evaluation, meaning values are computed only when needed. This can save a lot of processing time if not all values are required.
* Regular functions calculate everything upfront before returning a result, which can be slower if you only need a subset of the data.

Example:

```python
# Generator that stops processing as soon as the result is found
def find_first_even():
    for i in range(1000000):
        if i % 2 == 0:
            yield i

first_even = next(find_first_even())  # Stops as soon as it finds the first even number
```

c.  Represent Infinite Sequences
* Generators can represent infinite sequences, like an unbounded stream of numbers or lines from an open-ended log file, because they only compute values as needed.
* Regular functions cannot handle infinite sequences because they require a complete data structure to return.

Example:

```python
def infinite_counter():
    n = 0
    while True:
        yield n
        n += 1

counter = infinite_counter()
for _ in range(5):
    print(next(counter))
```

d. Improved Performance for Large Data Processing
* Generators can improve performance when processing large datasets because they allow for pipelining of data (processing elements one by one). This enables you to handle each element as soon as it’s ready, without waiting for the whole dataset to be prepared.
* Regular functions require complete preparation before processing, leading to slower performance when working with large datasets.

Example:

```python
def read_large_file(filename):
    with open(filename) as file:
        for line in file:
            yield line.strip()  # Processing line by line
```

e. Simplified Code for Iterators
* Writing a generator is often simpler and cleaner than manually implementing an iterator using classes with iter() and next() methods.
* A generator handles all the state maintenance internally, whereas a custom iterator requires more boilerplate code.

Example:

```python
# Generator for squares
def square_numbers():
    for i in range(10):
        yield i * i

# Custom iterator class (equivalent functionality, more complex)
class SquareNumbersIterator:
    def __init__(self):
        self.n = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.n < 10:
            result = self.n * self.n
            self.n += 1
            return result
        else:
            raise StopIteration()
```

f. Enhanced Readability
* Generators often make the code more readable because they reduce the need for complex state management. The yield keyword clearly shows where values are produced.
* Regular functions, especially those involving iteration and complex logic, can become cluttered with manual state tracking.

g. Support for Pipeline Style Programming
* Generators support pipeline processing (similar to Unix pipes), allowing you to chain operations without building intermediate data structures.
* This reduces both memory usage and processing time.

9. Example the purpose and usage of the map() function in Python.

Ans :     

The map() function in Python is a built-in function that applies a specified function to each item of an iterable (like a list, tuple, or set) and returns an iterator containing the results. This makes it easy to transform data without explicitly writing loops.

Purpose of map()

The primary purpose of map() is to apply a function to every item in an iterable and generate a new iterable with the transformed data. This is especially useful when you want to perform an operation on a sequence of data, such as modifying or filtering it.

Syntax of map()


```python
# map(function, iterable, ...)
```

Parameters:
* function: A function that will be applied to each element in the iterable(s). It can be a user-defined function, a lambda function, or a built-in function.
* iterable: One or more iterables (like a list, tuple, or set) whose elements will be passed to the function.

Return Value:
* The map() function returns a map object, which is an iterator. It can be converted to a list, tuple, or set if needed.

Example

```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4, 5]

# Using map() to apply the square function to each number
squared_numbers = map(square, numbers)

# Converting the map object to a list
print(list(squared_numbers))
```

Using map() with Lambda Functions

You can use a lambda function directly inside map() to avoid defining a separate function:

```python
numbers = [1, 2, 3, 4, 5]
doubled_numbers = map(lambda x: x * 2, numbers)

# Converting to a list
print(list(doubled_numbers))
```

Using map() with Multiple Iterables

The map() function can also take multiple iterables as input. In this case, the function should take as many arguments as there are iterables. The iteration stops when the shortest iterable is exhausted.

```python
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

sum_numbers = map(lambda x, y: x + y, numbers1, numbers2)

# Converting to a list
print(list(sum_numbers))
```

Using map() for String Manipulation

The map() function can also be used to manipulate strings:

```python
names = ['alice', 'bob', 'charlie']
uppercase_names = map(str.upper, names)

# Converting to a list
print(list(uppercase_names))
```

Using map() with User-Defined Functions

You can also use map() with custom, user-defined functions:

```python
def add_prefix(name):
    return f"Dr. {name}"

names = ['John', 'Smith', 'Doe']
prefixed_names = map(add_prefix, names)

# Converting to a list
print(list(prefixed_names))
```

10. What is the difference between map(), reduce(), and filter() function in Python ?

Ans :    

a. map() Function

The map() function is used to apply a function to every item in an iterable and returns an iterator with the transformed items.

Key Points
* Purpose: To transform each element in an iterable by applying a given function.
* Input: A function and one or more iterables.
* Output: A map object (which is an iterator) containing the transformed items.
* Returns the same number of elements as the original iterable(s).

Syntax

```python
# map(function, iterable, ...)
```

Example

```python
# square each number in a list
numbers = [1, 2, 3, 4, 5]
squared = map(lambda x: x ** 2, numbers)
print(list(squared))
```

b. filter() Function

The filter() function is used to filter elements in an iterable based on a specified condition (function). It returns an iterator containing only the elements that satisfy the condition.

Key Points
* Purpose: To filter elements in an iterable based on a condition.
* Input: A function that returns True or False, and an iterable.
* Output: A filter object (which is an iterator) containing only the elements for which the function returns True.
* The output may have fewer elements than the original iterable.

Syntax

```python
# filter(function, iterable)
```

Example

```python
# get even numbers from a list
numbers = [1, 2, 3, 4, 5, 6]
evens = filter(lambda x: x % 2 == 0, numbers)
print(list(evens))
```

c. reduce() Function

The reduce() function is used to apply a function of two arguments cumulatively to the items of an iterable, reducing the iterable to a single accumulated value. Unlike map() and filter(), reduce() is not a built-in function in Python 3; it is part of the functools module.

Key Points
* Purpose: To aggregate or reduce an iterable to a single cumulative value using a function that takes two arguments.
* Input: A function that takes two arguments and an iterable.
* Output: A single value.
* Useful for summing, multiplying, or combining items.

Syntax

```python
# from functools import reduce
# reduce(function, iterable, initializer=None)
```

Example

```python
# calculate the product of all numbers in a list
from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```
